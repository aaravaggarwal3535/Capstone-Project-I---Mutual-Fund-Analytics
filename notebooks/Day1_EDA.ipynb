```python
import pandas as pd

nav_history = pd.read_csv('../data/raw/02_nav_history.csv')

# unique AMFI codes in NAV History
nav_unique_codes = nav_history[['amfi_code']].drop_duplicates()

# pandas merge left join
merged_df = pd.merge(fund_master, nav_unique_codes, on='amfi_code', how='left', indicator=True)

#find rows that only existed in fund_master table and not in nav_history table
missing_codes_df = merged_df[merged_df['amfi_code'].isnull()]

total_funds = len(fund_master)
missing_count = len(missing_codes_df)

# data quality report
print("DATA QUALITY REPORT")
print(f"Validation Task: Cross-referencing AMFI codes")
print(f"Total schemes in Fund Master: {total_funds}")
print(f"Total AMFI codes missing in NAV History: {missing_count}\n")

if missing_count == 0:
    print("all AMFI codes from the fund master exist in the NAV History.")
else:
    print(f"Found {missing_count} orphaned schemes in Fund Master.")
    print("Missing Codes:")
    print(missing_codes_df['amfi_code'].tolist())
```
